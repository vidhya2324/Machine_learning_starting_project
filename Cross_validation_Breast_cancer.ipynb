```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

#Data collection

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

* M-->malignant-->0
* B-->beningnant-->1

separating features and target

above in pandas libraries..now we convert into numpy

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
 X = np.asarray(X)
 Y = np.asarray(Y)
```

#Gridsearchcv

gridsearchcv is used to determine the best parameter

```python
model=SVC()
```

```python
#hyperparameter
parameter={'kernel':['linear','poly','rbf','sigmoid'],'C':[1,5,10,20]} #c means its doesn't allow the misclassification
```

```python
#grid search
classifier=GridSearchCV(model,parameter,cv=5)#cv is 5 means...5 folds
```

```python
#fitting the data
classifier.fit(X, Y)
```

```python
classifier.cv_results_ #its show the result about the spliting...its will automatically split into 5..because we give 5 fold
```

```python
best_parameter=classifier.best_params_
print(best_parameter)
```

```python
accuracy=classifier.best_score_
print(accuracy)
```

```python
#loading the result to pandas framework
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10, 'kernel':'linear'}

#**RandomizedSearchCV**

```python
#loading the model
model =SVC()
```

```python
# hyperparameters

parameters = {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
#randomizedsearchcv
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
best_parameter = classifier.best_params_
print(best_parameter)
```

```python
accuracy=classifier.best_score_
print(accuracy)
```

```python
result=pd.DataFrame(classifier.cv_results_)
```

```python
print(result)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
 randomized_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10, 'kernel':'linear'}
