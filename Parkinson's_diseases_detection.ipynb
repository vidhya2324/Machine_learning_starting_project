#data preprocessing

```python
#drop the name columns
dataset=dataset.drop(columns=["name"],axis=1)
```

```python
dataset.info()
```

```python
#distribution of status columns
dataset['status'].value_counts()
```

* 0-->affected by disease
* 1-->not affected


```python
#data based on target variable(status colunms)
dataset.groupby('status').mean()
```

#train and test split

```python
x=dataset.drop(columns=["status"],axis=1)
y=dataset["status"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=6)
```

In above, all columns are decimal ways..so we do the standardscaler. It could gives all the values in same range. but the meaning of the data which already specified in the colunms will not change. the values is same...but here we comes all the data in same range.

```python
#datastandard scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

#model building

```python
model=svm.SVC(kernel="linear")
```

```python
model.fit(x_train,y_train)
```

#model evoluation

```python
x_train_prediction=model.predict(x_train)
training_data_prediction= accuracy_score(y_train, x_train_prediction)
print("accuracy score: ", training_data_prediction)
```

```python
x_test_prediction=model.predict(x_test)
testing_data_prediction= accuracy_score(y_test, x_test_prediction)
print("accuracy score: ", testing_data_prediction)
```
