##Data preprocessing

```python
#know about the categorical data
print(Car_dataset.Fuel_Type.value_counts())
print(Car_dataset.Seller_Type.value_counts())
print(Car_dataset.Transmission .value_counts())
```

```python
#getting the categorical data
catcols=Car_dataset.select_dtypes(include='object').columns
```

```python
print(catcols)
```

```python
#Encoding the Fuel_Tyes
Car_dataset.replace({'Fuel_Types':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

#Encoding the Seller_Type
Car_dataset.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

#Encoding the Transmission
Car_dataset.replace({'Transmission':{'Manual':0, 'Automatic':1}},inplace=True)
```

```python
Car_dataset.head()
```

#Splitting the data into test and training dataset

```python
x=Car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
y=Car_dataset['Selling_Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=2)
```

##Linear regression

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#model evaluation
training_prediction=model.predict(x_train)
```

```python
#R_squared error
error_score=metrics.r2_score(y_train,training_prediction)
print("R_squared error score ", error_score)
```

```python
test_prediction=model.predict(x_test)
```

```python
# R_squared error
error_score=metrics.r2_score(y_test,test_prediction)
print("R_squared error score ",error_score)
```

##Visualize the actual vs predicted values

```python
plt.scatter(y_train,training_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs predicted values")
plt.show()
```

```python
#visualize for test data
plt.scatter(y_test,test_prediction)
plt.xlabel('Actual price')
plt.ylabel("predictive price")
plt.title("actual price vs predictive price")
plt.show()
```

##lasso regression model evaluation

```python
from sklearn.linear_model import Lasso
```

```python
model=Lasso()
```

```python
model.fit(x_train,y_train)
```

```python
training_data_prediction=model.predict(x_train)
```

```python
error_score=metrics.r2_score(y_train,training_data_prediction)
print("R squared error ",error_score)
```

```python
test_data_prediction=model.predict(x_test)
```

```python
error_score=metrics.r2_score(y_test,test_data_prediction)
print("R_squared error",error_score)
```

* For training data, linear regression give more accuary compare to lasso regression
* For test data, lasso regresssion give more accuary compare to linear regression
